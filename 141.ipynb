### **Cleaning Data**

```python
import numpy as np
import pandas as pd
import chardet
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from collections import Counter
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
path = "Fatal Police Shootings 2015-Present.csv"
```

```python
with open(path, "rb") as rawdata:
    result = chardet.detect(rawdata.read(100000))
print(result)
```

```python
df_raw = pd.read_csv(path, index_col="id", encoding="ISO-8859-1")
```

```python
df_raw.columns
```

```python
df_raw.isnull().sum()
```

- 6 Features have missing value (name, armed, age, gender, race, flee). In further steps it will be cleaned.

```python
df_raw.date = pd.to_datetime(df_raw['date'], format='%Y-%m-%d')
```

- Converting [date] columns to datetime

```python
check_tail = df_raw.tail()
```

- Checking last row of the dateframe. Row contains full data. Do not need to remove it.

```python
df_raw.gender.unique()
```

- One value was marked as nan, so we converted it to most common value.

```python
df_raw.gender.fillna("M", inplace=True)
df_raw.gender.unique()
```

- Replacing nan value with male "M". Because only one value were missing.

```python
df_raw.race.value_counts()
df_raw.race.fillna("Other", inplace=True)
df_raw.race.unique()
```

- Replace nan value with "Other" races.

```python
age_mean = round(df_raw.age.mean())
df_raw.age.fillna(age_mean, inplace=True)


def conver_float_to_int(*args:int)->int:
    '''
    df["age"] values converts from float64 to int64 values
    '''
    for age in df_raw.age.values:
        df_raw["age"] = df_raw["age"].astype('int64')
    return df_raw["age"]

conver_float_to_int()
```

- Replace nan values with mean of age values.

```python
df_raw.signs_of_mental_illness.unique()
```

- Checking unique values for columns [signs_of_mental_illness]. Two unique values: True and False.

```python
df_raw.flee.fillna("Other", inplace=True)
df_raw.flee.unique()
```

- Replacing nan values with "Other".
- Checking unique values for columns [threat_level]. Three unique values: ['Not fleeing', 'Car', 'Foot', 'Other']

```python
df_raw.body_camera.unique()
```

- Checking unique values for columns [body_camera]. Two unique values: True and False.

```python
df_raw.armed.fillna("unknown weapon",inplace=True)
```

- All nan values for columns [armed] replace with "unknown weapon".

```python
df_raw.name.fillna("unknown name", inplace=True)
```

- All nan values for columns [name] replace with "unknown name".

```python
df_raw.isna().sum()
```

- All nan values were replace. 

```python
df_raw = df_raw.apply(lambda x: x.astype(str).str.lower())
```

- Converting all rows data to lowercase. 

```python
df_raw.to_csv("Fatal Police Shootings 2015-Present ISO-8859-1.csv")
```

- Saving formated data to new file: "Fatal Police Shootings 2015-Present ISO-8859-1.csv"

## **Analysis**

```python
df = pd.read_csv("Fatal Police Shootings 2015-Present ISO-8859-1.csv", index_col="id", encoding="ISO-8859-1")
```

```python
df.columns
```

## Death by Gender

```python
death_by_gender = df.gender.value_counts()
plt.title("Death by Gender",fontsize=16)
labels=["male", "female"]
autopct=lambda p:f'{p:.2f}%'
plt.pie(death_by_gender, labels=labels,autopct=autopct, radius=1.3);
```

- **95.85%** men.
- **4.15%** women. 

```python
data = df[["gender","age"]]
data.columns = ["gender","age"]
plt.title("Age Distribution in Each Gender", fontsize=16);
ax = sns.violinplot(x="gender", y="age", data=data, palette=['b','#F08030'], scale="width", cut=0)
plt.yticks((np.arange(0, 100, step=10)))
plt.show()
```

- Looking into violinplot we can see that mans age are spread more, than womans age. 

## Death by Age

```python
df.age.min()
df.age.max()
df.age.mean()
```

- Youngest person shot by police was 6 years old.
- Oldest person shot by police was 89 years old.
- Average age of person shot by police is 36.4.

```python
age_list = df.age.to_list()
histogram = Counter(min(age // 10 * 10, 100) for age in age_list)

plt.bar([x + 5 for x in histogram.keys()],  
        histogram.values(),                 
        10,                                 
        edgecolor=(0, 0, 0))             

plt.xticks([10 * i for i in range(11)])    
plt.xlabel("Age", fontsize=16)
plt.ylabel("Deaths", fontsize=14)
plt.title("Death by Age", fontsize=14)
plt.show()
```

- Highest amount of shooting are within 20 - 40 years old group.

- From the chart we can see that most amount shot by race is in white race. Over **1000** cases. This date has to be divided by race population to see better results.
